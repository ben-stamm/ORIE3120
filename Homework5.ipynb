**QUESTION 1**

```python
dropoff = taxi['tpep_dropoff_datetime'].astype('datetime64')
pickup = taxi['tpep_pickup_datetime'].astype('datetime64')
duration = (dropoff-pickup).dt.total_seconds()
taxi['duration'] = duration
taxi
```

```python
taxi['rounded_trip_distance'] = taxi['trip_distance'].round(0)
```

```python
question1a = taxi[['trip_distance', 'rounded_trip_distance', 'duration']]
question1a # --ANSWER 1a
```

```python
question1b = question1a.groupby('rounded_trip_distance')['trip_distance','duration'].sum()
question1b['velocity'] = question1b['trip_distance'] / (question1b['duration'] / 3600)
```

```python
plot = question1b.plot(y='velocity', xlim = (0,50), ylim = (0,50), title = "HW5 Question 1b")
plot.set_xlabel("Distance (mi)")
plot.set_ylabel("Velocity (mi/h)")
plot # --ANSWER 1b
```

```python
question1c = question1a
velocities = question1a['trip_distance'] / (question1a['duration'] / 3600)
question1c['velocity'] = velocities
```

```python
q1c = question1c[question1c['velocity'] < 100]
q1c = q1c[q1c['duration'] > 0]
q1c.head()
```

```python
plot2 = q1c.plot.scatter(x='trip_distance', y='velocity', s=.005, xlim = (0,50), ylim = (0,50), title = "HW5 Q1c")
plot2.set_xlabel("Distance (mi)")
plot2.set_ylabel("Velocity (mi/h)")
plot2 # --ANSWER 1c
```

```python
velocity_average = q1c['velocity'].mean()
trip_distance_sum = q1c['trip_distance'].sum()
duration_sum = q1c['duration'].sum()
print(velocity_average, (trip_distance_sum/duration_sum) * 3600) # --ANSWER 1d
```

**QUESTION 3**

```python
# Plot velocity versus hour of day, filter to weekdays, duration < 1 hour
q3 = taxi
q3['hour of day'] = taxi['tpep_dropoff_datetime'].astype('datetime64').dt.hour
q3['day of week'] = taxi['tpep_dropoff_datetime'].astype('datetime64').dt.dayofweek
q3['duration_hours'] = q3['duration'] / 3600
q3 = q3[q3['duration_hours'] < 1]
q3 = q3[q3['day of week'] < 5]
```

```python
q3_data = q3[['duration_hours','hour of day', 'trip_distance']]
```

```python
q3_grouped = q3_data.groupby('hour of day')['trip_distance', 'duration_hours'].sum()
q3_grouped['velocity'] = q3_grouped['trip_distance'] / q3_grouped['duration_hours']
```

```python
plot = q3_grouped.plot(y='velocity', ylim = (8,25), title = "HW5 Question 1b")
plot.set_xlabel("Hour of Day")
plot.set_ylabel("Velocity (mi/h)")
plot # --ANSWER 1b
```

From 12AM - 5AM, the average velocity increases from 16mph to peak at ~24MPH at 5AM. 


It then quickly declines to 10mph around 8am, as the work day begins and drivers get going on the road and traffic increases. 


The average speed then hovers around 10mi/hr during the work day (roughly 8AM-6PM) as traffic stays relatively stable throughout the day. <br>


Then, as people leave work and the roads de-clutter (roughly 6PM-12AM), the average speed increases relatively quickly from 10MPH to a local maximum plateauing around 14MPH. This plateau can be explained by the stabilization of traffic during dinner time.<br>


It is finally important to note that there is a significant increase in the velocity from post dinner time
to 12AM, indicating that a significant magnitdue of cars get off the road after 'dinner time'
